```python
import pandas as pd
#Loading the dataset
file_path = '/content/diabetes.csv'
diabetes_data = pd.read_csv(file_path)
#First few rows of the dataset
diabetes_data.head()
```

```python
#Generating descriptive statistics of the dataset
diabetes_data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
#Distribution of Glucose Levels by Outcome
sns.histplot(data=diabetes_data, x="Glucose", hue="Outcome", kde=True, bins=20, ax=axes[0])
axes[0].set_title("Distribution of Glucose Levels by Diabetes Outcome")
axes[0].set_xlabel("Glucose Level")
axes[0].set_ylabel("Frequency")
#BMI vs Age with Outcome as Hue
sns.scatterplot(data=diabetes_data, x="Age", y="BMI", hue="Outcome", palette="coolwarm", ax=axes[1])
axes[1].set_title("BMI vs Age by Diabetes Outcome")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("BMI")
#Plots
plt.tight_layout()
plt.show()
```

```python
#Calculating the correlation matrix
correlation_matrix = diabetes_data.corr()
#Displaying the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Diabetes Dataset")
plt.show()
```
